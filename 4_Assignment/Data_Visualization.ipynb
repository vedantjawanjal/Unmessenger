# Data Cleaning

From the given dataset we can see that the Last_review and  reviews_per_month has missing values 
also the last_review column should be converted to datetime format.

```python
# Convert 'last_review' column to datetime
data1['last_review'] = pd.to_datetime(data1['last_review'])
data1.info()
```

```python
# Handle missing values
# You can choose to drop rows with missing values or fill them with appropriate values

data1 = data1.dropna(subset=['last_review', 'reviews_per_month'])
data1.isnull().sum()
```

```python
# Check for duplicates
duplicate_rows = data1[data1.duplicated()]

# Print duplicate rows
print("Duplicate Rows:")
print(duplicate_rows)
```

```python
data1.describe().T
```

# Visualization 1: Histogram of Prices

```python
plt.figure(figsize=(8, 6))
plt.hist(data1['price'], bins=10, color='skyblue')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Histogram of Prices')
plt.show()
```

```python
data1['price'].max()
```

```python
# Check the skewness of the 'price' column
price_skewness = data1['price'].skew()
print("Skewness of Price column:", price_skewness)
```

```python
# mean > median
mean = data1['price'].mean()
median = data1['price'].median()
print(f'mean {mean} is greater than median {median} it is postive skew' )
```

<h3> Interpretation:</h3>

Given the magnitude of the skewness value, it indicates that there are potentially some high-priced outliers or properties with significantly higher prices compared to the majority of the properties.

# Visualization 2: Bar Plot of Room Types

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=data1, x='room_type', palette='viridis')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.title('Bar Plot of Room Types')
plt.show()
```

```python
max_nights_by_room_type = data1.groupby('room_type')['minimum_nights'].max()
print(max_nights_by_room_type)

plt.figure(figsize=(8, 6))
max_nights_by_room_type.plot(kind='bar', cmap='Set3')
plt.xlabel('Room Type')
plt.ylabel('Maximum Night Spends')
plt.title('Maximum nights spend as per room types')
plt.show()
```

# Visualization 3: Box Plot of Prices by Neighbourhood Group

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data1, x='neighbourhood_group', y='price', palette='pastel')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Price')
plt.title('Box Plot of Prices by Neighbourhood Group')
plt.show()
```

# Visualization 4:  Box Plot of Prices by Room Type

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data1, x='room_type', y='price', palette='pastel')
plt.xlabel('Room Type')
plt.ylabel('Price')
plt.title('Box Plot of Prices by Room Type')
plt.show()
```

```python
# Extract month from the 'last_review' column
data1['review_month'] = data1['last_review'].dt.month

# Group by month and calculate average availability
availability_by_month = data1.groupby('review_month')['availability_365'].mean()

# Find the month with the lowest availability
lowest_availability_month = availability_by_month.idxmin()
lowest_availability_value = availability_by_month.min()

print("Month with the lowest availability:", lowest_availability_month)
print("Average availability in that month:", lowest_availability_value)
```

```python
# Plot the average availability by month
plt.figure(figsize=(10, 6))
availability_by_month.plot(marker='o', color='skyblue')
plt.xlabel('Month')
plt.ylabel('Average Availability')
plt.title('Average Availability by Month')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.show()
```

```python
# Group by neighbourhood_group and room_type and calculate average availability
availability_by_group = data1.groupby(['neighbourhood_group', 'room_type'])['availability_365'].mean().unstack()

# Plot the average availability by neighbourhood_group and room_type
plt.figure(figsize=(12, 8))
availability_by_group.plot(kind='bar', cmap='Set3')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Average Availability')
plt.title('Average Availability by Neighbourhood Group and Room Type')
plt.xticks(rotation=45)
plt.legend(title='Room Type')
plt.tight_layout()
plt.show()
```

# Visualization 5: Pie Chart of Room Types

```python
plt.figure(figsize=(8, 8))
room_type_counts = data1['room_type'].value_counts()
plt.pie(room_type_counts, labels=room_type_counts.index, autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'lightcoral', 'lightsalmon'])
plt.title('Pie Chart of Room Types')
plt.show()
```

# Visualization 6: Heatmap of Correlation

#  Distribution of Numerical Features

Let's visualize the distribution of numerical features such as 'Salary', 'EngagementSurvey', 'DaysLateLast30', 'Absences', 	'EmpSatisfaction' and 'SpecialProjectsCount'.

```python
# Creating subplots
fig, axes = plt.subplots(3, 2, figsize=(12, 10))

# Distribution of Salary
sns.histplot(data['Salary'], bins=20, kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Distribution of Salary')

# Distribution of Engagement Survey
sns.histplot(data['EngagementSurvey'], bins=20, kde=True, ax=axes[0, 1])
axes[0, 1].set_title('Distribution of Engagement Survey')

# Distribution of EmpSatisfaction
sns.histplot(data['EmpSatisfaction'], bins=5, kde=True, ax=axes[1, 0])
axes[1, 0].set_title('Distribution of Employee Satisfaction')

# Distribution of SpecialProjectsCount
sns.histplot(data['SpecialProjectsCount'], bins=10, kde=True, ax=axes[1, 1])
axes[1, 1].set_title('Distribution of Special Projects Count')

# Distribution of Days Late Last 30
sns.histplot(data['DaysLateLast30'], bins=10, kde=True, ax=axes[2,0])
axes[2,0].set_title('Distribution of Days Late Last 30')

sns.histplot(data['Absences'], bins=10, kde=True, ax=axes[2,1])
axes[2,1].set_title('Distrbution of Absences')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

```python
''' Frequency Distribution of Categorical Features : - RecruitmentSource, Position, Sex, MaritalDesc, CitizenDesc, 
HispanicLatino, RaceDesc, TermReason, EmploymentStatus, Deperatment, PerformanceScore   '''
```

```python
data.head()
```

```python
import warnings
warnings.filterwarnings("ignore")

categorical_columns = ['RecruitmentSource','Sex', 'MaritalDesc', 'CitizenDesc', 
                       'HispanicLatino', 'RaceDesc', 'TermReason', 'EmploymentStatus', 'Department', 
                       'PerformanceScore', 'Position']
# Set up subplots
num_cols = 2
num_rows = (len(categorical_columns) - 1) // num_cols + 1
fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 30))

# Flatten axes for easy iteration
axes = axes.flatten()

# Iterate over categorical columns and create count plots
for i, column in enumerate(categorical_columns):
    sns.countplot(data[column], ax=axes[i], palette='viridis')
    axes[i].set_title(f'Frequency Distribution of {column}', fontsize=14)
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count', fontsize=12)
    axes[i].tick_params(axis='x', rotation=90)  # Rotate x-axis labels
    axes[i].grid(axis='y')
    
# Remove empty subplots if any
for i in range(len(categorical_columns), len(axes)):
    fig.delaxes(axes[i])

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Selecting only numerical columns for correlation analysis
numerical_data = data.select_dtypes(include=['int64', 'float64'])

# Calculating Pearson correlation matrix
correlation_matrix = numerical_data.corr()

# Plot correlation matrix as a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
# Select pairs of numerical features for scatter plots
scatter_pairs = [('Salary', 'SpecialProjectsCount'), ('PerfScoreID', 'EngagementSurvey'),
                 ('PerfScoreID', 'EmpSatisfaction'), ('FromDiversityJobFairID','EmpStatusID'), ('Zip', 'DeptID'),
                 ('EmpSatisfaction', 'EngagementSurvey'), ('Salary', 'EngagementSurvey')]

# Set up subplots
num_plots = len(scatter_pairs)
num_cols = 2
num_rows = (num_plots - 1) // num_cols + 1
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 30))

# Flatten axes for easy iteration
axes = axes.flatten()

# Create scatter plots for each pair of features
for i, (x_feature, y_feature) in enumerate(scatter_pairs):
    sns.scatterplot(x=data[x_feature], y=data[y_feature], ax=axes[i])
    axes[i].set_title(f'Scatter Plot of {x_feature} vs {y_feature}')
    axes[i].set_xlabel(x_feature)
    axes[i].set_ylabel(y_feature)

# Remove empty subplots if any
for i in range(num_plots, len(axes)):
    fig.delaxes(axes[i])
```

```python
# Select categorical and numerical features for box plot analysis
categorical_feature = 'Department'  # Choose the categorical feature
numerical_feature = 'Salary'  # Choose the numerical feature

# Create a box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x=categorical_feature, y=numerical_feature, data=data)
plt.title('Box Plot of Salary Across Departments')
plt.xlabel('Department')
plt.ylabel('Salary')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)
plt.show()
```
